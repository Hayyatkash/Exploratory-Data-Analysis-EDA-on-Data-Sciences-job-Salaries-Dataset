# 5. Salary Analysis

### 5.1 Ploting 3 graphs that shows salary by work year


```python
fig, axes = plt.subplots(1,3,figsize=(15,5))

# plot on each subplot
sns.barplot(data=df,x='work_year',y='salary',ax=axes[0])
sns.scatterplot(data=df,x='work_year',y='salary', ax=axes[1])
sns.boxplot(data=df,x='work_year',y='salary', ax=axes[2])
# set tick label format to plain decimal notation for each subplot
sns.set_style('ticks')
for ax in axes.flat:
    ax.ticklabel_format(style='plain', axis='y')
plt.show()
```

### 5.2 Ploting 3 graphs that shows salary by experience level

```python

fig, axes = plt.subplots(1,3,figsize=(15,5))

# plot on each subplot
sns.barplot(data=df,x='experience_level',y='salary',ax=axes[0])
sns.scatterplot(data=df,x='experience_level',y='salary', ax=axes[1])
sns.boxplot(data=df,x='experience_level',y='salary', ax=axes[2])
# set tick label format to plain decimal notation for each subplot
sns.set_style('ticks')
for ax in axes.flat:
    ax.ticklabel_format(style='plain', axis='y')
plt.show()
```

### 5.3 Ploting 3 graphs that shows salary by company size 


```python
fig, axes = plt.subplots(1,3,figsize=(15,5))

# plot on each subplot
sns.barplot(data=df,x='company_size',y='salary',ax=axes[0])
sns.scatterplot(data=df,x='company_size',y='salary', ax=axes[1])
sns.boxplot(data=df,x='company_size',y='salary', ax=axes[2])
# set tick label format to plain decimal notation for each subplot
sns.set_style('ticks')
for ax in axes.flat:
    ax.ticklabel_format(style='plain', axis='y')
plt.show()
```

### 5.4 Ploting a graph that shows highest salaries by job title


```python
plt.figure(figsize=(10,7))
# count the occurrences of each item in the 'column' column
value_counts = df['salary'].value_counts()

# select the top 3 most common items
top_7 = value_counts.nlargest(7)

# create a new dataframe with only the top 3 items
df_top_7 = df[df['salary'].isin(top_7.index)]

# create a countplot of the top 3 most common items
sns.set(style="darkgrid")
ax=sns.countplot(x='salary',hue='job_title',data=df_top_7)
# get the legend from the plot
# move the legend to the left
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))

# set the font size of the legend to small
plt.setp(ax.get_legend().get_texts(), fontsize='small')

# show the plot
plt.show()
```

### 5.5 Ploting a graph that shows Average Salary by Company Location 


```python
plt.figure(figsize=(10,8))
# group the dataframe by 'category', and calculate the average price
avg_salary = df.groupby('company_location')['salary'].mean().round(2)

# reset the index to make 'category' a column again
avg_salary = avg_salary.reset_index()

sns.set(style="darkgrid")
# plot a bar chart using Seaborn
sns.barplot(x='salary', y='company_location', data=avg_salary)

# calculate mean, median, and standard deviation of a numeric variable
print('Mean:', df['salary'].mean())
print('Median:', df['salary'].median())
print('Standard Deviation:', df['salary'].std())
```
